```python
import pandas as pd
import numpy as np
import os
```

```python
file_to_load = os.path.join("Kickstarter_Challenge.xlsx")

kickstarter_df = pd.read_excel(file_to_load)
kickstarter_df
```

```python
# Tabulate data in "outcomes" column for "Theater Outcomes by Launch Date"
kickstarter_df["outcomes"].unique()
```

```python
kickstarter_df["outcomes_tabulated"] = np.random.randn(len(kickstarter_df))
kickstarter_df["outcomes_tabulated"]
```

```python
kickstarter_df["outcomes_tabulated"] = kickstarter_df["outcomes"].copy(deep=True)

outcomes_list = ["successful", "failed", "canceled"]
```

```python
kickstarter_df.loc[kickstarter_df["outcomes_tabulated"] ==  "successful"] = 1
kickstarter_df.loc[kickstarter_df["outcomes_tabulated"] ==  "failed"] = 2
kickstarter_df.loc[kickstarter_df["outcomes_tabulated"] ==  "canceled"] = 3
kickstarter_df.loc[~kickstarter_df["outcomes_tabulated"].isin(outcomes_list)] = 1

kickstarter_df # How did this happen
```

```python
for i in range(len(kickstarter_df)):
    if kickstarter_df.loc[i,"outcomes"] == "successful":
        kickstarter_df.at[i,"outcomes_tabulated"] = 1
    elif kickstarter_df.loc[i,"outcomes"] == "failed":
        kickstarter_df.at[i, "outcomes_tabulated"] = 2
    elif kickstarter_df.loc[i,"outcomes"] == "canceled":
        kickstarter_df.at[i,"outcomes_tabulated"] = 3
    else:
        kickstarter_df.at[i,"outcomes_tabulated"] = 0
```

```python
kickstarter_df["outcomes_tabulated"]
```

```python
# Create bins for "Outcomes Based on Goals"
goals = kickstarter_df['goal']
goal_bins = [0, 1000, 4999, 9999, 14999, 19999, 24999, 
    29999, 34999, 39999, 44999, 50000, 100000000]
goal_ranges = ['Less than 1k', '1k-5k', '5k-10k', '10k-15k', '15k-20k', 
    '20k-25k', '25k-30k', '30k-35k', '35k-40k', '40k-45k', 
    '45k-50k', '50k+']

kickstarter_df["Goal Ranges"] = pd.cut(goals, goal_bins, labels=goal_ranges)
```
